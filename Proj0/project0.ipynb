```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

c = 0.25
t = np.linspace(0, 50, 5000)  # time array

d_x = np.cos(0.1 * t)
d_y = np.sin(0.12 * t)
d_z = np.sin(0.08 * t)

Ap_x = c * np.cos(t) + np.cos(0.1 * t)
Ap_y = c * np.sin(t) + np.sin(0.12 * t)
Ap_z = np.sin(0.08 * t)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Robot position
ax.plot(d_x, d_y, d_z, label='Robot')

# A_p(t) position
ax.plot(Ap_x, Ap_y, Ap_z, label='p(t)')

# Labels and legend
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('robot position and A_p(t)')
ax.legend()

# Show plot
plt.show()
```

```python
# Define the gravity vector at t=1
gravity_vector_t1 = np.array([1.1, 0.9, -9.7])

# Define the rotation matrix R(t) as a function of t


def rotation_matrix(t):
    return np.array([
        [np.cos(t), -np.sin(t), 0],
        [np.sin(t), np.cos(t), 0],
        [0, 0, 1]
    ])


# Calculate the rotation matrix at t=1 and t=5
R_t1 = rotation_matrix(1)
R_t5 = rotation_matrix(5)

# The rotation matrix from t=1 to t=5 is the product of R(t=5) and the inverse of R(t=1)
# However, since R(t) is orthogonal, its inverse is equal to its transpose
rotation_change = R_t5.T @ R_t1

# Apply the rotation change to the gravity vector at t=1 to find the gravity vector at t=5
gravity_vector_t5 = rotation_change @ gravity_vector_t1

gravity_vector_t5
```
